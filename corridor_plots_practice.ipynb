```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import style
```

```python
# Read the csv created in notebook "corridors_census_2016.ipynb" and saved in "Resources" folder on my desktop computer
# Save the data from the csv in a variable, "data_2016"
data_2016 = "./CSVs/CommCorr_Census_Stats_2016.csv"
```

```python
# Create a dataframe ("df_2016") by reading the variable data from "data_2016"
df_2016 = pd.read_csv(data_2016)
df_2016
```

```python
# Add a column for the year "2016" and insert it in the dataframe
df_2016['Year']= "2016"
df_2016.head()
```

```python
# remove columns "Census_tract," "GEOID," "Median household income," "poverty rate," and "Unemployment rate"
corridors_2016 = df_2016.drop(["Census_tract", "Per capita income", "GEOID", "Median household income","Poverty rate", "Unemployment rate" ], axis=1)
corridors_2016.head()
```

```python
# Use .groupby an .agg to sum the amounts for each corridor so it is easier to run analysis
df_2016_statistics = corridors_2016.groupby(['Corridor'], as_index=False).agg(
    {'Population': sum, 'Poverty count': sum, '# employed, age 16+': sum,'Unemployment count':sum, '# persons age 25+ graduated high school': sum, 
     'Pop. Black': sum, 'Pop. 2 or more races': sum, 'Pop. Hispanic origin': sum, 'Pop. Asian': sum, 
     'Total pop. in occupied housing units by tenure': sum, 'Total owner-occupied units': sum, 'Total renter-occupied units': sum, 'Pop. <18 years': sum,
     'Pop. working age': sum, 'Pop. 65+ years': sum}
)
df_2016_statistics
```

```python
# Create a bar chart to show population of each corridor for 2016

corr = df_2016_statistics['Corridor']
pop = df_2016_statistics['Population']

fig = plt.figure(figsize=(10,7))

# Sort the bar chart from corridor with lowest population to highest
df_sorted = df_2016_statistics.sort_values('Population')
plt.bar('Corridor', 'Population', data=df_sorted, color = 'blue')

# Rotate labels of the x axis so they're easier to read
plt.xticks(rotation=70)

#x-axis labels 
plt.xlabel('Commercial Corridor') 

#y-axis labels 
plt.ylabel('Population') 

#plot title 
plt.title('Population by Corridor - 2016') 

plt.show()
```

```python
# Create a scatter plot to show poverty count and Black population of corridors
# Want to see if there is a positive, negative, or no correlation between the 2 variables; plot shows a positive correlation

corr = df_2016_statistics['Corridor']
pop = df_2016_statistics['Population']
povcount = df_2016_statistics['Poverty count']
Black = df_2016_statistics['Pop. Black']

sns.scatterplot(data=df_2016_statistics, x="Poverty count", y="Pop. Black", hue='Corridor')
plt.title("Black Population vs. Poverty Count 2016", size = 20)
plt.show()
```

```python
# Another way to create scatterplot
corr = df_2016_statistics['Corridor']
pop = df_2016_statistics['Population']
povcount = df_2016_statistics['Poverty count']
Black = df_2016_statistics['Pop. Black']

plt.scatter(povcount, Black, edgecolors='r')

#x-axis labels 
plt.xlabel('Poverty count') 

#y-axis labels 
plt.ylabel('Black Population') 

#plot title 
plt.title('Black Population vs. Poverty Count - 2016') 

plt.show()
```

```python
# Collect the mean unemployment rate for the corridors
avg_unemploy = unemployment_merge.mean()

# Collect the years where data was collected
years = avg_unemploy.keys()
```

```python
# Plot the corridor average as a line chart
corr_avg, = plt.plot(years, avg_unemploy, color="blue", label="Corridor Average")

# Plot the unemployment values for a single corridor
pim, = plt.plot(years, unemployment_merge.loc['Pimlico',["2016","2017","2018","2019","2020"]], 
                        color="green",label="Pimlico")

# Plot the unemployment values for a different corridor
hamp, = plt.plot(years, unemployment_merge.loc['Hampden',["2016","2017","2018","2019","2020"]], 
                        color="gray",label="Hampden")

# Create a legend 
plt.legend(handles=[corr_avg, pim, hamp], loc="best")

#plot title 
plt.title('Average Unemployment Rate') 
plt.tight_layout()

# Save graph as a png file
plt.savefig("avg_unemploy.png")

# Show the chart
plt.show()
```
